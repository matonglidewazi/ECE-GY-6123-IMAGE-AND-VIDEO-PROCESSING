```python
import os
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Loading the Image and Label
* using (nibabel)[https://nipy.org/nibabel/nibabel_images.html]
* the data path is img_xx
* the label path is bv_bodyxx

```python
# set up the image path
image_path = 'bv_body_data/predict/'
data_path = 'img_'
label_path = 'bv_body'
appendix_str = '.nii'

sample_index = 0

data_file = os.path.join(image_path, data_path + str(sample_index) + appendix_str)
label_file = os.path.join(image_path, label_path + str(sample_index)+ appendix_str)
```

```python
import nibabel as nib
data_0 = (nib.load(data_file)).get_fdata()
image_0 = (nib.load(label_file)).get_fdata()
```

```python
fig, ax = plt.subplots(1,2)

slicing_index = 90

ax[0].imshow(data_0[slicing_index], cmap='gray')
ax[1].imshow(image_0[slicing_index])
plt.show()
```
