```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
filename = 'checkpoint2019-04-27 17:33:25.995535.pth'

checkpoint = torch.load(filename, map_location=torch.device('cpu'))
logger = checkpoint['logger']
```

```python
trainloss = logger['train']
epochs = len(trainloss)
valloss = [logger['validation'][i].numpy() for i in np.arange(epochs)]

plt.plot (np.arange(epochs), np.log(trainloss))
plt.plot (np.arange(epochs), np.log(valloss))
plt.show()
```
