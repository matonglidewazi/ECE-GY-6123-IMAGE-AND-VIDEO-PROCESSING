```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
# filename = 'checkpoint2019-04-29 12:16:17.484157.pth'
# filename ='checkpoint2019-05-01 15:18:55.567120.pth'
# filename ='checkpoint2019-05-01 09:49:28.653642.pth'

# 150 epoch 0.11 +_ 0.01
# filename ='checkpoint2019-05-07 09:21:18.907596.pth'

# 100 epoch 0.13 +_ 0.014
# filename ='checkpoint2019-05-07 09:13:25.648356.pth'

# 50 epoch 0.09 +_ 0.004
# filename = 'checkpoint2019-05-07 16:51:16.729294.pth'

# 200 epoch 0.106 +_ 0.006
# filename = 'checkpoint2019-05-07 15:27:40.379437.pth'

# 350 epoch 0.093 +_ 0.007
filename = 'checkpoint2019-05-08 09:48:23.554126.pth'

checkpoint = torch.load(filename, map_location=torch.device('cpu'))
logger = checkpoint['logger']
```

```python
trainloss = logger['train']
epochs = len(trainloss)
valloss = [logger['validation'][i].numpy() for i in np.arange(epochs)]
plt.ylim(0.05,0.15)
plt.plot (np.arange(epochs), trainloss)
plt.plot (np.arange(epochs), valloss)
plt.show()
print('best validation loss is {} in epoch {}'.format(np.min(valloss), np.argmin(valloss)))
print('latest 25 epoch stats is +- {} {}'.format (np.std(valloss[-25:]), np.mean(valloss[-25:])))
```
